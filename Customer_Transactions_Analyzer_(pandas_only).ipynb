Time-Based Analytics

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Date'].dt.day
```

```python
df['Date'].dt.weekday
```

```python
#Resample transactions by day/week/month
df['Date'] = pd.to_datetime(df['Date'])
df_time = df.set_index('Date')
```

```python
#daily revenue
daily_revenue = df_time.resample('D')['Transaction Amount'].sum()   #('W') for weekly and ('M') for monthly
daily_revenue
```

Rolling computes metrics over a moving window.
Unlike resample, rolling does not collapse rows â€” it adds context to each row.
Rolling always needs sorted time order.

```python
#Compute rolling sums/averages per customer
df = df.sort_values(['Customer ID', 'Date'])
```

```python
df['rolling_7d_sum'] = (
    df.groupby('Customer ID')['Transaction Amount']
    .rolling(7)
    .sum()
    .reset_index(level=0, drop=True)  #Level 0 â†’ Customer ID; Level 1 â†’ original DataFrame index
)
```

Ranking

```python
#Rank transactions per customer
df.groupby('Customer ID')['Transaction Amount'].rank(method='dense',ascending=False)
```
