```python
from sklearn.datasets import load_breast_cancer
```

```python
#Predict wheter an entity is high risk of getting cancer
#Malignant = 0 - High Risk
#Benign = 1 - Low Risk
```

```python
df = load_breast_cancer()
df
```

```python
data = load_breast_cancer(as_frame=True).frame
data.head()
```

Load and Split data

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = load_breast_cancer(return_X_y=True)
```

```python
X
```

```python
y
```

Pre-Processing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)

for trainidx, testidx in split.split(X, y):
  X_train, X_test = X[trainidx], X[testidx]
  y_train, y_test = y[trainidx], y[testidx]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

Choose a model

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(X_train_scaled,y_train)
```

```python
clf.classes_
```

```python
y_pred = clf.predict(X_test_scaled)
y_pred
```

```python
#Probabilities (needed for ROC-AUC)
y_prob = clf.predict_proba(X_test_scaled)[:,1]
y_prob
```

Accuracy - how often am I right?

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

Recall - Did I catch the dangerous cases?

```python
from sklearn.metrics import recall_score
recall_score(y_test, y_pred, pos_label=0)   #0.97 means you caught 97% of malignant tumors
```

Why pos_label=0?
In this dataset:

0 = malignant (high risk)

1 = benign

We care about recall of malignant cases.

ROC-AUC = Can I rank risk correctly?

```python
from sklearn.metrics import roc_auc_score
auc=roc_auc_score(y_test, y_prob)
auc
```

ROC-AUC Curve

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

fpr, tpr,threshold = roc_curve(y_test, y_prob)

plt.plot(fpr,tpr,label=f"ROC AUC = {auc:.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

Try different thresholds

```python
import numpy as np
from sklearn.metrics import precision_score

for t in [0.9,0.7,0.5,0.3,0.1]:
  y_pred_t = (y_prob>=t).astype(int)
  recall = recall_score(y_test, y_pred_t)
  precision = precision_score(y_test, y_pred_t)
  print(f"Threshold {t}: Recall score {recall: .3f}, Precision score {precision: .3f}")
```
