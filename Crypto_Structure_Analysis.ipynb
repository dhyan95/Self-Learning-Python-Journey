Concatenate

```python
conc_df = pd.concat([aave_df,binance_df,bitcoin_df,cardano_df,chainlink_df,dogecoin_df,ethereum_df,litecoin_df,solana_df,tether_df,USDCoin_df,XRP_df],axis=0)
conc_df.shape
```

```python
conc_df.columns
```

```python
conc_df['Symbol'].value_counts()
```

Sort and Checks

```python
conc_df.sort_values(by=["Date","Name"],inplace=True)
conc_df
#The code conc_df.sort_values(by=["Date","Name"],inplace=True) sorts the conc_df DataFrame.
#It first sorts the data by the 'Date' column and then, for rows with the same date, it sorts by the 'Name' column.
#The inplace=True argument ensures that these changes are applied directly to the conc_df DataFrame, modifying it in place rather than returning a new sorted DataFrame
```

```python
print("Missing values per column:\n", conc_df.isnull().sum())
print("\nNumber of duplicate rows:", conc_df.duplicated().sum())
```

Price -> Return

```python
conc_df['Daily return'] = conc_df.groupby('Symbol')['Close'].pct_change()
conc_df[['Symbol','Name','Date','Volume','Daily return']].head(10)
#pct_change() calculates (today_close - yesterday_close) / yesterday_close.
```

Pivot Table

```python
pivot1 = conc_df.pivot(index='Date',columns='Symbol',values='Daily return')
pivot1.head(10)
```

```python
pivot1.shape
```

```python
conc_df.shape
```

```python
pivot1.isna().sum()
```

Drop NAN rows

```python
pivot_clean = pivot1.dropna(axis=0)
pivot_clean
```

```python
pivot_clean.shape
```

Convert Pivot Table to NumPy

```python
X = pivot_clean.values
X
```

```python
X.shape
```

Standardization

```python
mean = X.mean(axis=0)  #each coin gets its own mean
mean
```

```python
std = X.std(axis=0)    #each coin gets its own std
std
```

```python
stand = (X - mean) / std    #each coin is computed separately
stand
```

```python
mean.shape
```

```python
std.shape
```

```python
stand.shape
```

Matrix Operation

```python
#corrcoef
corr = np.corrcoef(stand.T)  #rowvar=False is equivalent to stand.T
corr
#rowvar=False: This argument tells np.corrcoef that the variables (cryptocurrencies in this case) are arranged in columns, not rows.
#So, it calculates the correlation between columns.
```

Visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f",xticklabels=pivot_clean.columns, yticklabels=pivot_clean.columns)
plt.title('Correlation Heatmap of Daily Returns')
plt.show()
```
