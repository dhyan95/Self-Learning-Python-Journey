Sales Data Analysis
You have daily sales data for a small store for a month (30 days) stored as a list:

```python
sales = [200, 220, 250, 180, 210, 240, 230, 260, 200, 190, 220, 210, 250, 270, 300, 310, 290, 280, 260, 270, 300, 320, 310, 300, 290, 280, 270, 260, 250, 240]

#Convert this list into a NumPy array.
sales_arr = np.array(sales)
print("Sales Array:", sales_arr)

#Compute the average sales, maximum sales, and minimum sales.
print("Average sales:", np.average(sales_arr))
print("Maximum sales:", np.max(sales_arr))
print("Minimum sales:", np.min(sales_arr))

#Find the days when sales exceeded 280.
high_sales_days = sales_arr[sales_arr > 280]
print('Sales that exceeded 280:', high_sales_days)
```

Temperature Monitoring
A city records hourly temperatures in Celsius over a week (7 days  24 hours = 168 readings) as a NumPy array called temps.

```python
#Convert the temperatures to Fahrenheit.
temps_cel = np.random.rand(168) + 30
temps_cel

temps_fah = (temps_cel - 30)*2
temps_fah

#Identify the coldest day and the hottest day based on daily averages.
print("Coldest day in fahrenheit, my arse is freezing: ",np.min(temps_fah))
print("Hottest day in fahrenheit: ",np.max(temps_fah))

#Find all hours when the temperature was below 0C.
below = temps_fah[temps_fah<0.6]
print("Hours when the temperature was below 0C: ",below)
```

Class Grades
You have grades for 5 students in 4 subjects as a NumPy array:

```python
grades = np.array([
 [85, 90, 78, 92],
 [70, 75, 80, 72],
 [88, 82, 91, 85],
 [60, 65, 58, 62],
 [95, 92, 88, 96]
])

#Compute each students average grade.
student1_average = np.average(grades[0])
print("Student 1: ",student1_average)
student2_average = np.average(grades[1])
print("Student 2: ",student2_average)
student3_average = np.average(grades[2])
print("Student 3: ",student3_average)
student4_average = np.average(grades[3])
print("Student 4: ",student4_average)

#Find the subject with the highest average score.
subject1_average = np.average(grades[:,0])
print("Subject 1: ",subject1_average)

#Identify students scoring above 90 in at least 2 subjects.
above_90_counts = np.sum(grades > 90, axis=1)
students_above_90 = np.where(above_90_counts >= 2)[0]
print("Students scoring above 90 in at least 2 subjects:", students_above_90)
```

```python
import numpy as np
np.dot([3j+0,5j+2],[2j+5,1j+2])
```

Vdot - same syntax as dot, unlike dot it flattens the array into a 1D array and then performs the operations.
Not as powerful as vecdot

Multi Dot

```python
from numpy.linalg import multi_dot
A = np.random.rand(100,100)
B = np.random.rand(100,10)
C = np.random.rand(10,100)
multi_dot([A,B,C])
```

VecDot - working with multiple arrays and when one wants to use broadcasting functions on arrays. Very similar to linalg.vecdot - for specifying explicitly that this is a linear algebra function

```python
v = np.array([(0., 5., 0.), [0., 0., 10.], [0., 6., 8.]])
n = np.array([0., 0.6, 0.8])
np.vecdot(v, n)
```

Tensor dot

```python
a_0 = np.array([[1, 2], [3, 4]])
b_0 = np.array([[5, 6], [7, 8]])
c_0 = np.tensordot(a_0,b_0,axes=0)
c_0
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and a 3D axes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Generate data for the mesh grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Mesh Grid Plot')

plt.show()
```
